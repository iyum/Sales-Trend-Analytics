Find weekly sales

```python
# Group by 'week' and calculate the total sales for trend analysis
weekly_sales_trend = task1_cleaned_df.groupby('week')['Sales'].sum().reset_index()

# Display the weekly sales trend
weekly_sales_trend
```

Find anomali from sales data

```python
import matplotlib.pyplot as plt

# Plotting the weekly sales trend to visually inspect for anomalies
plt.figure(figsize=(10, 6))
plt.plot(weekly_sales_trend['week'], weekly_sales_trend['Sales'], marker='o')
plt.title('Weekly Sales Trend')
plt.xlabel('Week')
plt.ylabel('Total Sales')
plt.grid(True)
plt.show()
```

I analyzed the provided dataset to identify sales trends on a weekly basis. The total sales for each week were calculated, revealing a significant drop in sales during Week 18
